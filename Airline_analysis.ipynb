Problem Statement: You are working for an airline company that wants to analyze its passenger data for the past year. The company has provided you with a CSV file containing the monthly passenger data for each airline. Your task is to use NumPy, Pandas, and Matplotlib to analyze the data and answer the following questions:

1) What is the trend of monthly passenger traffic for each airline? Plot the line graph to show the same.

2) What is the overall monthly passenger traffic trend for all airlines combined? Plot the line graph to show the same.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("flights.csv")
data.head(15)
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
lista_meses = data["month"][0:12]
meses = pd.api.types.CategoricalDtype(categories=lista_meses)
data["month"] = data["month"].astype(meses)
```

```python
data.groupby(["year", "month"])["passengers"].mean().unstack().plot(kind="line", marker = "o")
plt.title("Passengers Over Time")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend(title = "Month", bbox_to_anchor = (1.05,1), loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```

```python
data.groupby("month")["passengers"].mean().plot(kind="line", marker = "o")
plt.title("Passengers Over Time")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend(title = "Month", bbox_to_anchor = (1.05,1), loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```
